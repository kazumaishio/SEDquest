## 開発
### - 事象生成(Gauss関数の確率分布)
### - Gauss fit to the histogram
- これに倣った。

https://www.youtube.com/watch?v=ajMidM4AZSs

- 正規分布でのフィットの場合は単純

https://www.geeksforgeeks.org/how-to-plot-normal-distribution-over-histogram-in-python/

```python
import matplotlib.pyplot as plt
import numpy as np
def gaussian(x, amplitude, mu, sigma):
    # return (1.0 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2) : Normal distribution
    return amplitude * np.exp(-0.5 * ((x - mu) / sigma) ** 2)


x = np.random.normal(170,10,250)
print(x)

#############################################
#  Fit a Gaussian distribution to the data
#############################################
from scipy.optimize import curve_fit
nbins = 30
n, bins, patches = plt.hist(x, bins = nbins)
# plt.show()

# Fit a gaussian distribution to the data:
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, nbins)
y = n
popt, pcov = curve_fit(gaussian, x, y, p0=[1, np.mean(x), np.std(x)])
# popt, pcov = curve_fit(gaussian, x, y)

plt.plot(x, gaussian(x, *popt))#, 'k', linewidth=2)
print('amplitude', popt[0])
print('mu', popt[1])
print('sigma', popt[2])

# import numpy as np
# from scipy.stats import norm
# mu, std = norm.fit(x) 
# xmin, xmax = plt.xlim()
# x = np.linspace(xmin, xmax, 100)
# p = norm.pdf(x, mu, std)
# plt.plot(x, p, 'k', linewidth=2)
# title = "Fit Values: {:.2f} and {:.2f}".format(mu, std)
# plt.title(title)
# plt.show()
```

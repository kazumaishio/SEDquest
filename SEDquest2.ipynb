```python



def get_consecutive_bins(s,nconsecutive ):

  for i in range(0,len(s)-nconsecutive):
      if np.all(s[i:i+nconsecutive] > 2):
          print(i)
          print(s[i:i+nconsecutive])
          print('---')
          return i
      
  return -1
    
import numpy as np

nobs = 100000
nbins = 11
nconsecutive = 3
sample_size = nobs*nbins


mu, sigma = 0, 1.0
s = np.random.normal(mu, sigma, sample_size)

import matplotlib.pyplot as plt
# fig = plt.figure()

n_detected_peaks = 0
for i,j in zip(range(0,sample_size,nbins), range(nobs)):
    # print(i,j)
    # print(s[i:i+nbins])
    peakbininit = get_consecutive_bins(s[i:i+nbins],nconsecutive)
    if peakbininit > 0:
        print('peakbininit',peakbininit)
        # plt.plot(range(peakbininit, nbins),s[i+peakbininit:i+nbins], label='obs'+str(j), marker='o',linestyle='')
        plt.plot(range(peakbininit, peakbininit+nconsecutive),s[i+peakbininit:i+peakbininit+3], label='obs'+str(j), marker='o',linestyle='-')
        n_detected_peaks += 1

    # plt.plot(range(nbins),s[i:i+nbins], label='obs'+str(j))

print('n_detected_peaks',n_detected_peaks)
print('nobs',nobs)
print('probability',n_detected_peaks/nobs)
# s[]



```

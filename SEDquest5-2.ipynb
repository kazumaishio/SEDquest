### ピーク検出・有意度計算

```python
################################################
### # Peak significance calculation - not used
### This must be used with the listing of all the data points
### of consecutive positive deviations, 
### which can be more than three
################################################
def calc_peak_significance(ri):
#     '''
#     fit (array) values for the fit
#     x,y,yerr (arrays) data
#     N total number of points
#     n_free number of parameters we are fitting
#     '''
#     return 1.0/(N-n_free)*sum(((fit - y)/yerr)**2)
# import numpy as np
# from scipy.stats import chi2, norm
# # only those residuals that form the (at least) 3 consecutive positive deviations
# ri = np.array([1, 2, 3.])

# # Chi-square of those residuals
  chi2_value = np.sum(ri**2)

# # p-value for chi-square with m degrees of freedom - m = len(ri) - this is the probability of obtaining a test statistic at least as high as chi2_value
  p_val = 1 - chi2.cdf(chi2_value, df=len(ri))

# # p-value to Gaussian-equivalent one-sided significance
  sigma_equiv = norm.ppf(1 - p_val)

  return p_val, sigma_equiv


###################
# Peak detection
###################
def get_consecutive_bins_for_multi_peaks(s,nconsecutive = 3 ):
  # from s, find all the sets of consecutive bins that have values > 1.0 for at least nconsecutive bins
  # return the list of (peakinitbin, peaklength, significance)s
  peakinitbins=[]
  peaklengths=[]
  significances=[]
  peakinitbin = -1
  for i in range(0,len(s)-nconsecutive+1):  
    if np.all(s[i:i+nconsecutive] > 1.0):
      if peakinitbin < 0:
        peakinitbin = i
      if i + nconsecutive == len(s):
        peakinitbins.append(peakinitbin)
        peaklengths.append(i + nconsecutive - peakinitbin)
        significances.append(calc_peak_significance(s[peakinitbin:i+nconsecutive]))
        return   peakinitbins,peaklengths,significances
    else:
      if peakinitbin >= 0:
        peakinitbins.append(peakinitbin)
        peaklengths.append(i + nconsecutive -1 - peakinitbin)
        significances.append(calc_peak_significance(s[peakinitbin:i+nconsecutive-1]))
        peakinitbin = -1
  return   peakinitbins,peaklengths,significances

def get_consecutive_bins(s,nconsecutive = 3 ):
  peakinitbin = -1
  for i in range(0,len(s)-nconsecutive+1):
    # print(i)
    if np.all(s[i:i+nconsecutive] > 1.0):
      #   print(i)
      #   print(s[i:i+nconsecutive])
      #   print('---')
      if peakinitbin < 0:
        peakinitbin = i
        # print('Found peak at bin#:',peakinitbin)
        # print(s[peakinitbin:i+nconsecutive])
      if i + nconsecutive == len(s):
        # print('Last bin reached', i, 'for the length ',len(s))
        # print('End of peak at bin #:',i + nconsecutive -1, 'length ', i + nconsecutive - peakinitbin)
        # print(s[peakinitbin:i+nconsecutive])
        # print(peakinitbin, i+ nconsecutive - peakinitbin)
        return peakinitbin, i+ nconsecutive - peakinitbin, calc_peak_significance(s[peakinitbin:i+nconsecutive])
    else:
      if peakinitbin >= 0:
        # print('End of peak at bin#:',i + nconsecutive -2, 'length ', i -1 + nconsecutive - peakinitbin)
        # print(s[peakinitbin:i+nconsecutive ])
        # print(peakinitbin, i+ nconsecutive -1 - peakinitbin)
        return peakinitbin, i+ nconsecutive -1 - peakinitbin, calc_peak_significance(s[peakinitbin:i+nconsecutive-1])
        # if peakbininit[0] > 0:
        #   print('Found longer peak at bin:',peakbininit[0] , 'in total length of ', len(s))
        # else :
        #   print('Longest peak at bin:',i ,'length ', nconsecutive, 'in total length of ', len(s))
  return -1, 0
```

## Get the table of analysis results on each obsdate (observation period)

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table, vstack
from astropy.modeling.powerlaws import LogParabola1D
from astropy.modeling import models, fitting
from astropy import units as u
from scipy.optimize import curve_fit



nbinsmin=9

###################
# Utility functions
###################
def gaussian(x, amplitude, mu, sigma):
    # return (1.0 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2) : Normal distribution
    return amplitude * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

def calc_reduced_chi_square(fit, x, y, yerr, N, n_free):
    '''
    fit (array) values for the fit
    x,y,yerr (arrays) data
    N total number of points
    n_free number of parameters we are fitting
    '''
    return 1.0/(N-n_free)*sum(((fit - y)/yerr)**2)


# eval_spectra from SEDquest5-1.ipynb
def eval_spectra(filepath): #'data/3C454.3_allsed_14d_min11.ecsv'
  # logpar_init = LogParabola1D(amplitude=1,x_0=10,alpha=1,beta=1)
  logpar_init = LogParabola1D(
    amplitude = 0.000016494113774149846,
    x_0 = 9.482855296727278,
    alpha = -0.5677548858840729,
    beta = 0.1071613245072173)

  ### initialize a linear fitter ###
  # fit = fitting.TRFLSQFitter()
  fit = fitting.DogBoxLSQFitter()

  ### Read the data ###
  t0 = Table.read(filepath)
  nonzero_mask = (t0['e2dnde'] > t0['e2dnde_err'])
  t = t0[nonzero_mask]


  ### MJDごとに処理 ###
  obsdates=np.unique(t['tstart'].data).tolist()
  n_detected_peaks = 0
  array_nbins = []
  # t_residual = Table()
  t_fitparams = Table()
  for idx, obsdate in enumerate(obsdates):
    # print(idx, ': obsdate',obsdate)
    mask = (t['tstart']==obsdate)
    x = t[mask]['e_ref']
    y = t[mask]['e2dnde']
    yerr = t[mask]['e2dnde_err']
    nbins = len(x)
    if nbins < nbinsmin:
        print('nbins < nbinsmin:',nbins)
        continue
    array_nbins.append(nbins)

    ### fit the data  ###
    try: 
        fitted_line = fit(logpar_init, x,y,weights=1.0/yerr, maxiter=200)
    except:
        print(f"!!!!!!!!! Fit failed for obsdate: {obsdate}")
        continue
    reduced_chi_squared = calc_reduced_chi_square(fitted_line(x), x, y, yerr, len(x), 4)
    if reduced_chi_squared > 4.0:
      # refit with another beta 
      logpar_init.beta.value = 0.1
      ## Implement the fit with error handling. If the fit fails, we can skip this obsdate and move on to the next one.
      try:
          fitted_line = fit(logpar_init, x,y,weights=1.0/yerr, maxiter=200)
          residuals = (y-fitted_line(x))/yerr
          reduced_chi_squared = calc_reduced_chi_square(fitted_line(x), x, y, yerr, len(x), 4)
      except:
          print(f"!!!!!!!!! Fit failed for obsdate: {obsdate}")
          continue

    residuals = (y-fitted_line(x))/yerr
    # residuals = (fitted_line(x)-y)/yerr # negative:dip finding
    # peakbininit = get_consecutive_bins(residuals,nconsecutive)
    nconsecutive = 3
    peakinitbins,peaklengths,significances = get_consecutive_bins_for_multi_peaks(residuals,nconsecutive)
    # t_residual_ind = Table([x, residuals,np.full(len(x),obsdate)],
    #                         names=[ "e_ref", "residual", "obsdate"])#"e_ref",x, "residual", residuals, "e2dnde", y, "e2dnde_err", yerr)
    # t_residual = vstack([t_residual,t_residual_ind])

    # Extract y value where x (e_ref) is closest to 100 MeV
    idx_100mev = np.argmin(np.abs(x - 100.0))
    y_at_100mev = y[idx_100mev]

    # Extract y value where x (e_ref) is closest to 200 MeV
    idx_200mev = np.argmin(np.abs(x - 200.0))
    y_at_200mev = y[idx_200mev]
    
    # NOTE: Wrap scalar values in lists for Table creation
    ispeak = 0
    if len(peakinitbins) > 0:
      ispeak = 1
    t_fitparams_ind = Table(
        [[obsdate], [nbins], [fitted_line.amplitude.value], [fitted_line.x_0.value], 
         [fitted_line.alpha.value], [fitted_line.beta.value], 
         [ispeak],
         [y_at_200mev], [y_at_100mev]],
        names=["obsdate", "nbins","amplitude", "x_0", "alpha", "beta", 
               "ispeak",
               "y_200MeV", "y_100MeV"])
    t_fitparams = vstack([t_fitparams, t_fitparams_ind])

    # if len(peakinitbins) > 0:
    #   # print('peakbininit',peakbininit)
    #   n_detected_peaks += 1
    #   plotSED(x, y , yerr, fitted_line,  peakinitbins, peaklengths, residuals, reduced_chi_squared, obsdate, idx)
    #   print('curvature beta:', fitted_line.beta.value)
    #   for peakbininit, peaklength in zip(peakinitbins, peaklengths):
    #     print(residuals[peakbininit:peakbininit+peaklength])
  return t_fitparams
```

```python
from astropy.table import Table
import matplotlib.pyplot as plt
import os
from astropy.table import Table, join

lc_file_dir = "./data_LC/14d"
t_analysis_summary = Table(names=['source_index', 'source_name', 'source_class', 
                                  'correlation_coefficient',
                                  'mean_beta', 'std_beta', 'mean_flux', 'std_flux'], 
                                  dtype=['i4', 'S20', 'S10', 'f8', 'f8', 'f8', 'f8', 'f8'])
for source in dict_sourcename[0:20]:
  print(source)
  lc_tab = Table.read(os.path.join(lc_file_dir, '4fgl_' + source['4FGL name'] + '_lightcurve.fits'))
  lc_tab.add_column((lc_tab['tmax_mjd'] + lc_tab['tmin_mjd']) / 2., name='obsdate', index=0)

  t_fitparams = eval_spectra(f'./data/{source["assoc name"]}_allsed_14d_min11.ecsv')

  ### Correlation between curvature beta and flux
  # make a table with columns obsdate, beta, flux
  # then 
  # - plot obsdate vs beta and obsdate vs flux in the same plot with two y-axes
  # - plot correlation between beta and flux with a scatter plot, and calculate the correlation coefficient

  # ### Table with columns obsdate, beta, flux
  # # The keys: 
  lc_tab['obsdate_for_join'] = lc_tab['tmin_mjd'].round(3) # round to 3 decimal places to avoid floating point issues
  t_fitparams['obsdate_for_join'] = t_fitparams['obsdate'].round(3) # round to 3 decimal places to avoid floating point issues
  # print('lc_tab ')
  # print(lc_tab['obsdate_for_join'])
  # print('t_fitparams ')
  # print(t_fitparams['obsdate_for_join'])
  t_joined = join(lc_tab, t_fitparams, keys='obsdate_for_join')# join_type='inner')
  
  ### Calculate correlation coefficient between curvature beta and flux
  corr_coeff = np.corrcoef(t_joined['beta'], t_joined['flux'])[0, 1]
  print(f'Correlation coefficient between curvature beta and flux: {corr_coeff}')
  ### Calculate mean and standard deviation of beta and flux
  mean_beta = np.mean(t_joined['beta'])
  std_beta = np.std(t_joined['beta'])
  mean_flux = np.mean(t_joined['flux'])
  std_flux = np.std(t_joined['flux'])

  t_analysis_summary.add_row([source['index'], source['assoc name'], source['CLASS'],
                              corr_coeff,
                              mean_beta, std_beta, mean_flux, std_flux])

  ### Drawing from SEDquest4_PKS0426.ipynb
  # obsdate vs curvature beta ###
  fig, ax = plt.subplots()
  ax.plot(t_fitparams['obsdate'], t_fitparams['beta'], '.', label='β')
  ax.set_xlabel('Observation date (MJD)')
  ax.set_ylabel('Curvature β')

  ax2 = ax.twinx()
  ax2.plot(lc_tab['obsdate'], lc_tab['flux'], 'r.', label='flux')
  ax2.set_ylabel('flux (cm$^{-2}$ s$^{-1}$)')
  # combine legends from both axes
  lines, labels = ax.get_legend_handles_labels()
  lines2, labels2 = ax2.get_legend_handles_labels()
  ax.legend(lines + lines2, labels + labels2, loc='best')
  plt.title(f'{source["index"]}: {source["assoc name"]}, class: {source["CLASS"]}')


  ### Correlation between curvature beta and flux ###
  # fig, ax = plt.subplots()
  # ax.plot(t_joined['beta'], t_joined['flux'], 'o')
  # ax.set_xlabel('Curvature β')
  # ax.set_ylabel('Flux (cm$^{-2}$ s$^{-1}$)')
  # plt.title(f'{source["index"]}: {source["assoc name"]}, class: {source["CLASS"]}')
  fig, ax = plt.subplots()
  ax.plot(t_joined['beta'], np.log10(t_joined['flux']), 'o')
  ax.set_xlabel('Curvature β')
  ax.set_ylabel('Log Flux (cm$^{-2}$ s$^{-1}$)')
  plt.title(f'{source["index"]}: {source["assoc name"]}, class: {source["CLASS"]}')

# print summary: correlation_coefficients
#   For all sources, 
print(t_analysis_summary)
```

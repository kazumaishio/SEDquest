```python

####################################################################
# for Inline animations in Jupyter
# https://stackoverflow.com/questions/43445103/inline-animations-in-jupyter
####################################################################
# Enable interactive plot
%matplotlib notebook

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import os

####################################################################
# Functions
####################################################################
##################################
# 
##################################
def met_to_mjd(met) :
    mjd = 51910. + 7.4287e-4 + met/86400.
    return mjd
# print(mjd_start, mjd_end)
# print(met_to_mjd(mjd_start), met_to_mjd(mjd_end))
##################################
#  GET fermi SED table from fits file
##################################
from astropy.table import Table
def getsed(timebin_start,timebin_end):
    # print(directories)
    directoryname=f'lightcurve_{timebin_start}_{timebin_end}'
    sed_path=os.path.join(input_directory_parent,directoryname,sed_filename )
    try : 
        sed_tab= Table.read(sed_path)
        # sed_tab= Table.read(f'/home/kazuma/Workspace/AGN_data/3_3C454_100MeV_14d/14days/{directoryname}/4fgl_j2253.9+1609_sed.fits')
        # sed_tab= Table.read(f'/home/kazuma/Workspace/AGN_data/3_3C454_100MeV_7d/7days/{directoryname}/4fgl_j2253.9+1609_sed.fits')
        return sed_tab
    except:
        print('not found: ', sed_path)    
        return None

from astropy import units as u
from astropy.io import ascii
```

```python

mjd_min=45000#55000
mjd_max=65000#56000

##################################
# Get subdirectories within time scope
directories=os.listdir(input_directory_parent)
# directories=os.listdir('/home/kazuma/Workspace/AGN_data/3_3C454_100MeV_7d/7days')
timebins_start=[]
timebins_end=[]
for directoryname in directories:
    elements=directoryname.split('_')
    timebin_start=met_to_mjd(float(elements[1]))
    timebin_end=met_to_mjd(float(elements[2]))
#     print(elements[1],elements[2])
#     print(timebin_start, timebin_end)
    if timebin_end>mjd_max*1.0001 :
        continue        
    if timebin_start<mjd_min*0.9999 :
        continue
    timebins_start.append(int(elements[1]))
    timebins_end.append(int(elements[2]))

timebins_start.sort()
timebins_end.sort()
print(timebins_start[0], timebins_end[-1])
print('Which are {} and {}'.format(mjd_min,mjd_max))
print(met_to_mjd(timebins_start[0]), met_to_mjd(timebins_end[-1]))


    
```

```python
##################################
# Get Fermi SED
##################################
n_significant_sed=0
for i in range(len(timebins_start)):
    timebin_start=timebins_start[i]
    timebin_end=timebins_end[i]
    sed_tab=getsed(timebin_start,timebin_end)
    if sed_tab is None:
        continue

    default_eref=sed_tab['e_ref']
    default_flux=np.ones(len(default_eref))*1e-4
    default_asymmerr=[np.ones(len(default_eref))*1e-5,np.ones(len(default_eref))*1e-5]

    default_errdiff=np.ones(len(default_eref))
    default_ts=np.ones(len(default_eref))*0.1
    high_significance_mask = (sed_tab['ts']>=9)
    low_significance_mask = (sed_tab['ts']<9)
    sed_tab['ShowUL']=low_significance_mask
    sed_tab_significant=sed_tab[high_significance_mask]
    sed_tab_nosignificant=sed_tab[low_significance_mask]
    print(len(sed_tab['e_ref']),len(sed_tab_significant['e_ref']))
    if len(sed_tab_significant['e_ref'])>=10:
        n_significant_sed+=1
    # print(len(sed_tab_nosignificant['e_ref']),len(sed_tab_nosignificant['e2dnde_ul']))
    # print(sed_tab_nosignificant['e_ref'],sed_tab_nosignificant['e2dnde_ul'])

print('Number of significant SEDs:',n_significant_sed)
```

```python
##################################
# Get Fermi SED
##################################

from astropy.table import Table, vstack
table_allSEDs=Table()
n_significant_sed=0
for i in range(len(timebins_start)):
    timebin_start=timebins_start[i]
    timebin_end=timebins_end[i]
    sed_tab=getsed(timebin_start,timebin_end)
    if sed_tab is None:
        continue

    default_eref=sed_tab['e_ref']
    default_flux=np.ones(len(default_eref))*1e-4
    default_asymmerr=[np.ones(len(default_eref))*1e-5,np.ones(len(default_eref))*1e-5]

    default_errdiff=np.ones(len(default_eref))
    default_ts=np.ones(len(default_eref))*0.1
    high_significance_mask = (sed_tab['ts']>=9)
    low_significance_mask = (sed_tab['ts']<9)
    sed_tab['ShowUL']=low_significance_mask
    sed_tab_significant=sed_tab[high_significance_mask]
    sed_tab_nosignificant=sed_tab[low_significance_mask]
    print(len(sed_tab['e_ref']),len(sed_tab_significant['e_ref']))
    if len(sed_tab_significant['e_ref'])>=10:
        n_significant_sed+=1
        sed_tab_significant['tstart']=[met_to_mjd(timebin_start)]*len(sed_tab_significant['e_ref'])
        sed_tab_significant['tend']=[met_to_mjd(timebin_end)]*len(sed_tab_significant['e_ref'])
        table_allSEDs=vstack([table_allSEDs, sed_tab_significant])

    # print(len(sed_tab_nosignificant['e_ref']),len(sed_tab_nosignificant['e2dnde_ul']))
    # print(sed_tab_nosignificant['e_ref'],sed_tab_nosignificant['e2dnde_ul'])


table_allSEDs.write(outfilename)
# print(t)

# print('Number of significant SEDs:',n_significant_sed)
```

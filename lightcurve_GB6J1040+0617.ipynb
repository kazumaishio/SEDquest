```python
from astropy.table import Table
import numpy as np


# pre-update (1d)
lc_tab0=Table.read(f'data_LC/1d/4fgl_j1040.5+0617_lightcurve.fits')
lc_tab0.add_column((lc_tab0['tmax_mjd'] + lc_tab0['tmin_mjd'])/2., name='t_mjd', index=0)

# post-update (3d)
lc_tab=Table.read(f'data_LC/GB6J1040+0617/4fgl_j1040.5+0617_lightcurve_3d.fits')
lc_tab.add_column((lc_tab['tmax_mjd'] + lc_tab['tmin_mjd'])/2., name='t_mjd', index=0)

# post-update (3d, 2nd version)
lc_tab2=Table.read(f'data_LC/GB6J1040+0617/4fgl_j1040.5+0617_lightcurve_3d_2.fits')
lc_tab2.add_column((lc_tab2['tmax_mjd'] + lc_tab2['tmin_mjd'])/2., name='t_mjd', index=0)
import matplotlib.pyplot as plt

plt.errorbar(lc_tab0['t_mjd'],lc_tab0['flux'], yerr=lc_tab0['flux_err'], 
                 label='pre-update (1d)',fmt='o', markersize=3, capsize=2, elinewidth=1, color='red')
plt.errorbar(lc_tab['t_mjd'],lc_tab['flux'], yerr=lc_tab['flux_err'], 
                 label='post-update1 (3d)',fmt='o', markersize=3, capsize=2, elinewidth=1, color='blue')
plt.errorbar(lc_tab2['t_mjd'],lc_tab2['flux'], yerr=lc_tab2['flux_err'], 
                 label='post-update2 (3d)',fmt='o', markersize=3, capsize=2, elinewidth=1, color='green')
plt.legend()

plt.xlabel('MJD')
plt.ylabel('flux (cm$^{-2}$ s$^{-1}$)')
plt.show()


# fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(15,10))

# err=np.ones(len(sed_tab_nosignificant))*sed_tab_nosignificant['e2dnde_ul']*0.5
# axes[1].errorbar(sed_tab_nosignificant['e_ref'], sed_tab_nosignificant['e2dnde_ul'] ,
#                  yerr=err,
#                  uplims=True,
#                  linestyle='None',color='red',ecolor='red',capsize=3)

# plt.plot(lc_x,lc_tab['flux_ul95'] ,marker='o', label = 'UL')
# lightcurve,= plt.plot(lc_x[fermipresentbin],lc_y[fermipresentbin] ,marker='o',color='red')
# plt.xlim(56895, 56901)

# plt.plot(lc_x,lc_y ,marker='o')

# plt.yscale('log')
# plt.ylim(1e-9,1e-5)
# plt.xlim(56890, 56910)
# plt.ylim(1e-13,1e-4)
# plt.xlim(56900, 56903)
# plt.xlim(56901.3, 56901.7)
# plt.ylim(1e-13,1e-2)
# plt.xlim(56901.2, 56901.7)
# plt.ylim(1.2e-16,1.4e-2)
# plt.xlim(56901.45, 56901.5)
# plt.ylim(1.2e-4,1)
# plt.legend()
```
